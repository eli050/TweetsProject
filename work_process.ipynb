```python
sum_of = 0
count = 0
for val in df['Text']:
  if val:
    sum_of += len(val.split())
    count += 1
print(sum_of /count)
```

```python
df["count_word"] = df['Text'].map(lambda x: len(str(x).split()))
df["count_word"]
```

```python
df["count_letters"] = df["Text"].map(lambda x: len(str(x)))
```

```python
from pprint import pprint
dict_df = dict()
for val in df['Biased'].unique():
  dict_df[str(val)] = df[df['Biased'].astype(str) == str(val)].sort_values(by="count_letters",ascending=False).head(3)['Text'].to_dict()
pprint(dict_df)
```
